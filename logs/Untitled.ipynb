```python
import pandas as pd
```

```python
frames = []

values = ["0.2", "0.4", "0.6", "0.8", "0.9", "0.95", "0.98", "0.99", "0.999"]

for val in values:
    frames.append(pd.read_csv("%sprofit.csv" % val))

df = pd.concat(frames, keys=values)
```

```python
df['discount'] = df.index
df['discount'] = df['discount'].apply(lambda x: x[0]).astype(float)
df
```

```python
import seaborn as sns

dots = sns.load_dataset("dots")


# Plot the lines on two facets
sns.relplot(
    data=df,
    x="km", y="reward",
    hue="discount",
    height=5, aspect=3,
)
```
